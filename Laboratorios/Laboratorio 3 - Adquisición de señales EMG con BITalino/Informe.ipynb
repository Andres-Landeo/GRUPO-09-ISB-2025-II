```python
import numpy as np
import matplotlib.pyplot as plt

# Definir frecuencia de muestreo
Fs = 1000   # Hz

# Cargar datos
emg = np.loadtxt("Señales Bitalino/EMG_brazo_final.txt")

# Vector de tiempo completo
t = np.arange(len(emg)) / Fs  

plt.figure(figsize=(12, 8))

# Prueba 1: Reposo
inicio = int(0 * Fs)
fin    = int(15 * Fs)
plt.subplot(3, 1, 1)
plt.plot(t[inicio:fin], emg[inicio:fin], color="navy")
plt.title("Prueba 1: Reposo")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# Prueba 2: Flexión
inicio = int(15 * Fs)
fin    = int(30 * Fs)
plt.subplot(3, 1, 2)
plt.plot(t[inicio:fin], emg[inicio:fin], color="darkgreen")
plt.title("Prueba 2: Flexión")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# Prueba 3: Flexión contra fuerza
inicio = int(30 * Fs)
fin    = len(emg)
plt.subplot(3, 1, 3)
plt.plot(t[inicio:fin], emg[inicio:fin], color="darkred")
plt.title("Prueba 3: Flexión contra fuerza")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

plt.tight_layout()
plt.show()
```
