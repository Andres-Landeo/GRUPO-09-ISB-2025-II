## **5.1. Bíceps braquial**

### **5.1.1. Prueba 1: Reposo**

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Parámetros y carga
Fs = 1000  # Hz
emg1 = np.loadtxt("Señales Bitalino/EMG_brazo_final.txt", usecols=[-1])
t = np.arange(len(emg1)) / Fs

# Reposo
inicio = int(0 * Fs); fin = int(10 * Fs)
x = emg1[inicio:fin]
tt = t[inicio:fin]

# Rectificar y Welch
x_dc = x - x.mean()        # quitar DC
x_rect = np.abs(x_dc)      # rectificación
nper = min(1024, len(x_dc))
f, Pxx = welch(x_dc, fs=Fs, nperseg=nper)  # Welch

plt.figure(figsize=(12, 9))

# 1) Señal cruda
plt.subplot(3, 1, 1)
plt.plot(tt, x, color="navy")
plt.title("Prueba 1: Reposo - EMG cruda")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# 2) Rectificada
plt.subplot(3, 1, 2)
plt.plot(tt, x_rect, color="darkgreen")
plt.title("EMG rectificada (|x - media|)")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# 3) Welch
plt.subplot(3, 1, 3)
plt.semilogy(f, Pxx, color="darkred")
plt.title("Espectro de potencia (Welch)")
plt.xlabel("Tiempo (s) / Frecuencia (Hz)")
plt.ylabel("Densidad espectral")
plt.grid(True)

plt.tight_layout()
plt.show()
```

### **5.1.2. Prueba 2: Flexión**

```python
# Prueba 2: Flexión
inicio = int(10 * Fs)
fin    = int(30 * Fs)

plt.figure(figsize=(12, 4))
plt.plot(t[inicio:fin], emg1[inicio:fin], color="darkgreen")
plt.title("Prueba 2: Flexión - Señal EMG cruda")
plt.ylabel("Amplitud (µV)")
plt.grid(True)
plt.show()
```

### **5.1.3. Prueba 3: Flexión contra fuerza**

```python
# Prueba 3: Flexión contra fuerza
inicio = int(30 * Fs)
fin    = int(40 * Fs)

plt.figure(figsize=(12, 4))
plt.plot(t[inicio:fin], emg1[inicio:fin], color="darkred")
plt.title("Prueba 3: Flexión contra fuerza - Señal EMG cruda")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (µV)")
plt.grid(True)
plt.show()
```
