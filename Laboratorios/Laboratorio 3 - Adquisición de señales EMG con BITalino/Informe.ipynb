## **5.1. Bíceps braquial**

### **5.1.1. Prueba 1: Reposo**

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir frecuencia de muestreo
Fs = 1000   # Hz

# Cargar datos
emg1 = np.loadtxt("Señales Bitalino/EMG_brazo_final.txt")

# Vector de tiempo completo
t = np.arange(len(emg1)) / Fs  

plt.figure(figsize=(12, 8))

# Prueba 1: Reposo
inicio = int(0 * Fs)
fin    = int(10 * Fs)
plt.plot(t[inicio:fin], emg1[inicio:fin], color="navy")
plt.title("Prueba 1: Reposo")
plt.ylabel("Amplitud (µV)")
plt.grid(True)
plt.show()
```

### **5.1.2. Prueba 2: Flexión**

```python
# Prueba 2: Flexión
inicio = int(10 * Fs)
fin    = int(30 * Fs)
plt.figure(figsize=(12, 8))
plt.plot(t[inicio:fin], emg1[inicio:fin], color="darkgreen")
plt.title("Prueba 2: Flexión")
plt.ylabel("Amplitud (µV)")
plt.grid(True)
plt.show()
```

### **5.1.3. Prueba 3: Flexión contra fuerza**

```python
# Prueba 3: Flexión contra fuerza
inicio = int(30 * Fs)
fin    = int(40 * Fs)
plt.figure(figsize=(12, 8))
plt.plot(t[inicio:fin], emg1[inicio:fin], color="darkred")
plt.title("Prueba 3: Flexión contra fuerza")
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud (µV)")
plt.grid(True)
plt.show()
```
