# **1. Introducción**
La electromiografía (EMG) es una técnica de diagnóstico usada para evaluar el estado y la salud de los músculos y los nervios que los controlan, nervios motores. Esta prueba detecta la actividad eléctrica producida por las fibras musculares en respuesta a la estimulación nerviosa, lo cual permite identificar disfunciones musculares, trastornos neuromusculares o problemas de comunicación entre nervios y músculos [1]. Dentro de un contexto de investigación, el EMG nos permite armar diferentes niveles de análisis, intengrandose de manera facil con tareas cognitivas o reflejos para mapear de manera correcta los estímulos presentes; de esta manera evita el sesgo del participante, convirtiendo sensaciones en data más confiable para un posterior análisis más objetivo, puesto que cada milisegundo se va detectando distintos cambios de valencia. Finalmente, la sencillez en el uso y la versatilidad de la prueba, la hacen muy didactica e ideal para pruebas que necesiten alta reproductibilidad como las clínicas. [2]

# **7. Anexo**

```python
# Ejemplo de código para análisis de señal EMG: 
# Abductor del pulgar: Prueba 3 - Abducción contra fuerza
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import welch

# Parámetros y carga
Fs = 1000  # Hz
emg = np.loadtxt("Señales Bitalino/EMG_pulgar_final.txt", usecols=[-1])
t = np.arange(len(emg)) / Fs

# Reposo (0-10 s)
# Movimiento (10-30 s)
# Movimiento contra fuerza (30-40 s)
inicio = int(30 * Fs)
fin    = int(40 * Fs)
x = emg[inicio:fin] # Movimiento contra fuerza
tt = t[inicio:fin]

# Rectificar y Welch
x_dc = x - x.mean()        # quitar DC
x_rect = np.abs(x_dc)      # rectificación
nper = min(1024, len(x_dc))
f, Pxx = welch(x_dc, fs=Fs, nperseg=nper)  # Welch

plt.figure(figsize=(12, 9))

# 1) Señal cruda
plt.subplot(3, 1, 1)
plt.plot(tt, x, color="navy")
plt.title("Prueba 3: Abducción contra fuerza - EMG cruda")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# 2) Rectificada
plt.subplot(3, 1, 2)
plt.plot(tt, x_rect, color="darkgreen")
plt.title("EMG rectificada")
plt.ylabel("Amplitud (µV)")
plt.grid(True)

# 3) Welch
plt.subplot(3, 1, 3)
plt.semilogy(f, Pxx, color="darkred")
plt.title("Espectro de potencia (Welch)")
plt.xlabel("Tiempo (s) / Frecuencia (Hz)")
plt.ylabel("Densidad espectral")
plt.grid(True)

plt.tight_layout()
plt.show()
```
